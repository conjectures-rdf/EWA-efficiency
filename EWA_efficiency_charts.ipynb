```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def log_line_charts(data, label, y_label):
  index = ['D1', 'D2', 'D3']

  df = pd.DataFrame(data, index=index)

  # Create a figure and axis
  fig, ax = plt.subplots()

  # Define line styles and colors
  line_styles = ['--', '-.', ':', '--', '-.', '--']
  line_colors = ['#EF563B', '#636EFA', '#AB63FA', '#FFA15B', '#00CC96', '#56C2D8']
  marker_styles = ['^', 'o', 's', 'x', 'v', 'D']

  # Plot each line with different style and color
  for i, (column, style, color, marker) in enumerate(zip(df.columns, line_styles, line_colors, marker_styles)):
      ax.plot(df.index, df[column], linestyle=style, color=color, marker=marker, markersize=8, label=column)


  # Add legend
  ax.legend()
  ax.set_yscale('log')


  # Set labels and title
  ax.set_xlabel('Dataset')
  ax.set_ylabel(y_label)
  ax.set_title(label)

  # Display the plot
  plt.show()
```

```python
def draw_barchart(data, threshold, label):
  line_colors = ['#EF563B', '#636EFA', '#AB63FA', '#FFA15B', '#00CC96', '#56C2D8']

  # Set up the figure and axis
  #fig, ax = plt.subplots(figsize=(6, 6))

  # Plotting each bar with specified colors
  bar_width = 0.15
  index = np.arange(len(data['queries']))
  opacity = 1

  for i, (method, values) in enumerate(data.items()):
      if method == 'queries':
          continue
      plt.bar(index + bar_width * i, values, bar_width, alpha=opacity, label=method, color=line_colors[i-1])

  # Adding labels and title
  plt.ylim(top=threshold)
  plt.xlabel('Queries')
  plt.ylabel('Execution time (ms)')
  plt.title(label)
  plt.xticks(index + bar_width * (len(data) - 1) / 2, data['queries'])
  #plt.legend()
  # plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=6)


  # Display the plot
  #plt.tight_layout()
  plt.show()
```
